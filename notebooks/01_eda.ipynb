```python
import pandas as pd

df = pd.read_csv("../data/raw.csv")

df.head()
```

```python
print(df.shape)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["abstract_length"] = (
    df["abstract"]
    .str.split()
    .apply(len)
)
```

```python
df["abstract_length"].describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(df["abstract_length"], bins=50)
plt.xlabel("Words")
plt.ylabel("Count")
plt.title("Abstract Length Distribution")
plt.show()
```

```python
df = df.drop_duplicates()

df = df.dropna(
    subset=["title", "abstract"]
)
```

```python
df.to_csv(
    "../data/clean_papers.csv",
    index=False
)
```
